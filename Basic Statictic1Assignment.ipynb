```python
##Descriptive Analytics
import pandas as pd
df=pd.read_csv("sales_data_with_discounts.csv")
print("Mean:\n",df.mean(numeric_only=True))
print("Median:\n",df.median(numeric_only=True))
print("Mode:\n",df.mode(numeric_only=True).iloc[0])
print("Standard Deviation:\n",df.std(numeric_only=True))

```

```python
##Data Visualization
##Histograms
import matplotlib.pyplot as plt
df.hist(figsize=(10,8))
plt.show()
```

```python
##Boxplot
import matplotlib.pyplot as plt
df.boxplot(figsize=(10,8))
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].value_counts().plot(kind='bar', title=col)
    plt.show()
```

```python
##Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df.select_dtypes(include=['float64', 'int64']))
scaled_df = pd.DataFrame(scaled, columns=df.select_dtypes(include=['float64', 'int64']).columns)

print(scaled_df.head())
```

```python
##Convert categorical Data to number
df_encoded = pd.get_dummies(df, drop_first=True)
print(df_encoded.head())
```
