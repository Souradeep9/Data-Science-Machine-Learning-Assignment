```python
##Dimensionality Reduction with PCA
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
```

```python
features = df.select_dtypes(include=[np.number])
```

```python
features
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
```

```python
scaled_data
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data
```

```python
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
```

```python
pca_result
```

```python
##Clustering with Original Data
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans
```

```python
clusters_orig = kmeans.fit_predict(scaled_data)
```

```python
clusters_orig
```

```python
sns.scatterplot(x=features.iloc[:, 0], y=features.iloc[:, 1], hue=clusters_orig)
plt.title("KMeans Clustering (Original Data)")
plt.show()
```

```python
print("Silhouette Score:", silhouette_score(scaled_data, clusters_orig))
```

```python
print("Davies-Bouldin Index:", davies_bouldin_score(scaled_data, clusters_orig))
```

```python
 ##Clustering with PCA Data
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca
```

```python
clusters_pca = kmeans_pca.fit_predict(pca_result)
clusters_pca
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=clusters_pca)
plt.title("KMeans Clustering (PCA Data)")
plt.show()
```

```python
print("Silhouette Score (PCA):", silhouette_score(pca_result, clusters_pca))
```

```python
print("Davies-Bouldin Index (PCA):", davies_bouldin_score(pca_result, clusters_pca))
```

##Conclusion and Insights
PCA helps with visualization and efficiency, especially for high-dimensional data.

Use clustering directly when feature relationships are strong and meaningful.

Use PCA before clustering when you want to:

Speed up processing

Remove redundancy

Visualize clusters in 2D or 3D
